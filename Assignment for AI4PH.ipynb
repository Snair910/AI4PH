# Heart Disease Dataset Analysis 
Data from: [Heart Disease Dataset on Kaggle](https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset/data)

## Context:
This dataset dates from 1988 and includes four databases: Cleveland, Hungary, Switzerland, and Long Beach V. It comprises 76 attributes, with the predicted attribute being the presence of heart disease in the patient. The target field is integer-valued, where 0 indicates no disease, and 1 indicates disease.

## Content:
### Attribute Information:
- age
- sex
- chest pain type (4 values)
- resting blood pressure
- serum cholesterol in mg/dl
- fasting blood sugar > 120 mg/dl
- resting electrocardiographic results (values 0, 1, 2)
- maximum heart rate achieved
- exercise-induced angina
- oldpeak = ST depression induced by exercise relative to rest
- the slope of the peak exercise ST segment
- number of major vessels (0-3) colored by fluoroscopy
- thal: 0 = normal; 1 = fixed defect; 2 = reversible defect

Note: The names and social security numbers of the patients were recently removed from the database and replaced with dummy values.

## Correlation Matrix

To explore relationship and patterns to look at potential dependencies

```python
# Correlation matrix
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

Higher chest pain type, maximum heart rate, and a specific slope type are positively correlated with a higher likelihood of heart disease.Exercise-induced angina (exang) and ST depression induced by exercise relative to rest (oldpeak) exhibit negative correlations with heart disease, suggesting that individuals with exercise-induced angina and higher ST depression are less likely to have heart disease.
Age (age): Negatively correlated with the target. Older age is associated with a lower likelihood of heart disease.
Sex (sex): Negatively correlated with the target. Being male is associated with a lower likelihood of heart disease.

```python
# Calculate absolute correlation coefficients with the target variable
correlations = df.corr()['target'].abs()

# Sort the correlations in descending order
sorted_correlations = correlations.sort_values(ascending=False)
print("Absolute Correlation Coefficients with the Target Variable:")
print(sorted_correlations)
```

By finding the absolute correlation coefficients, I wanted to look at just the strength of the relationship between each variable and the response without cosndiering the direction. Additionally, I calculated the mean,variance and std dev of the three highest predictor variables to understand them better.

```python
# Calculate mean, variance, and standard deviation for each variable
oldpeak_mean = df['oldpeak'].mean()
oldpeak_variance = df['oldpeak'].var()
oldpeak_std_dev = df['oldpeak'].std()

exang_mean = df['exang'].mean()
exang_variance = df['exang'].var()
exang_std_dev = df['exang'].std()

cp_mean = df['cp'].mean()
cp_variance = df['cp'].var()
cp_std_dev = df['cp'].std()

print(f"Oldpeak - Mean: {df['oldpeak'].mean()}, Variance: {df['oldpeak'].var()}, Standard Deviation: {df['oldpeak'].std()}")
print(f"Exang - Mean: {df['exang'].mean()}, Variance: {df['exang'].var()}, Standard Deviation: {df['exang'].std()}")
print(f"CP - Mean: {df['cp'].mean()}, Variance: {df['cp'].var()}, Standard Deviation: {df['cp'].std()}")
```

### Distribution of Age in the full dataset and in data of people with heart disease

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Plotting the distribution for the full data
sns.histplot(df['age'], bins=20, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Full Data - Age Distribution', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Age', fontsize=14)
axes[0].set_ylabel('Frequency', fontsize=14)

# Plotting the distribution for target 1
sns.histplot(df[df['target'] == 1]['age'], bins=20, kde=True, color='salmon', ax=axes[1])
axes[1].set_title('People with Heart Disease - Age Distribution', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Age', fontsize=14)
axes[1].set_ylabel('Frequency', fontsize=14)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

The age distribution analysis reveals interesting patterns in the dataset concerning the presence of heart disease (target variable 1). In the full dataset, there is a notable concentration of individuals around the age of 60, indicating a significant representation of this age group. This may suggest that individuals around 60 years old contribute substantially to the overall dataset. On the other hand, when focusing on the subset where heart disease is present (target 1), there is a distinct peak in frequency shortly after the age of 40. This finding implies that individuals in their 40s are more prevalent among those diagnosed with heart disease. The shift in age distribution between the full dataset and the target 1 subset suggests that age could be a crucial factor in understanding and predicting the occurrence of heart disease, with a specific emphasis on individuals in their 40s.

### Box plot for maximum heart rate by chest pain type

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='cp', y='thalach', data=df, palette='pastel')
plt.title('Boxplot for Maximum Heart Rate (thalach) by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

Chest Pain Type 0 (cp=0):
The range of maximum heart rates is more spread out, indicating higher variability.
Median heart rate for this type is lower than the medians of the other types.Indicates that individuals with Chest Pain Type 0 tend to have lower maximum heart rates compared to other chest pain types.

Chest Pain Type 1 (cp=1):
The range of maximum heart rates is more concentrated, indicating lower variability.
Median heart rate for this type is higher than the medians of the other types.Indicates that individuals with Chest Pain Type 1 tend to have higher maximum heart rates compared to other chest pain types.
